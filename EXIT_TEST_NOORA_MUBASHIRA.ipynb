NULL VALUE FILLIMG

````python
for col in ['ball_controlling_skills' ,'jumping_skills','penalties_conversion_rate','mental_strength','shot_accuracy','behaviour_rating','matches_played','fitness_rating','years_of_experience']:data[col]=data[col].fillna(data[col].median())
````

````python
data['weight'].mode()
````

````python
data['weight']=data['weight'].fillna(data['weight'].mode()[0])
````

````python
data['strong_foot']=data['strong_foot'].fillna(data['strong_foot'].mode()[0])
data['gender']=data['gender'].fillna(data['gender'].mode()[0])
data['coaching']=data['coaching'].fillna(data['coaching'].mode()[0])
````

````python
data["height"] = data["height"].apply(lambda x: int(x.split("'")[0]) * 12 + int(x.split("'")[1]))
data["weight"] = data["weight"].str.extract("(\d+)").astype(float)
````
